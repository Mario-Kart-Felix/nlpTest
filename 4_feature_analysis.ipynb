```python
featureText = open("title_features.txt", "r").read()
features =  featureText.split(",")
```

```python
import string
# from nltk.stem import WordNetLemmatizer
# wnLemm = WordNetLemmatizer()

from nltk.stem import PorterStemmer
from nltk.tokenize import sent_tokenize, word_tokenize

ps = PorterStemmer()

def cleanText(textList):
    cleanList = []
    for word in textList:
        stripped = word.strip()
        no_punct = "".join([x for x in stripped if x not in string.punctuation])
        cleaned = ps.stem(no_punct)
        if len(cleaned)>0:
            cleanList.append(cleaned)
    return cleanList
```

```python
feature_cleaned = cleanText(features)
```

```python
len(feature_cleaned)
```

```python
feature_cleaned[0:50]
```

```python
from collections import defaultdict

word_count = defaultdict(int) # default int value is 0
for word in feature_cleaned:
    word_count[word] += 1
    
print(word_count)
```

```python
import pandas as pd

dataDict = {"feature":list(word_count.keys()), "count": list(word_count.values())}

data = pd.DataFrame(dataDict)
```
