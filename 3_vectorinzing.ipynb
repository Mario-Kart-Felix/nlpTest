# vectorizing 

```python
# importing Data
f_title = open("dataContent.csv", "r").readlines()
print("Total rows in data = {}".format(len(f_title)))

# Cleaning up the escape charaters
rawData = [ x.strip() for x in f_title if len(x)>0]
print(rawData[0:5])
```

```python
import pandas as pd

data = pd.DataFrame(rawData, columns=['text'])
print(data[0:5])
```

```python
import string
import re
#import nltk

#nltk.download('wordnet')
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer

wnLemm = WordNetLemmatizer()
```

```python
def cleanData(text):
    '''
     one big cleaning function with five not best practice 
     functions : remove punctuation, tokenize, lower, remove stopwords, stemming
    '''
    # remove punctuations
    text_no_punct1 = "".join([x for x in text if x not in string.punctuation])
    # remove digits
    text_no_punct = "".join([x for x in text if x not in string.digits])
    # tokenize the text
    text_tokenized = re.split('\W+', text_no_punct)
    # convert to lowercase
    text_lower = [x.lower() for x in text_tokenized]
    # remove the stopwords
    text_no_stop = [x for x in text_lower if x not in stopwords.words('english')]
    # stemming the words
    #text_stemmed = [ps.stem(x) for x in text_no_stop]
    # picking instead lemmatizer for the text
    text_lemmetized = [wnLemm.lemmatize(x) for x in text_no_stop]
    # clean empty items
    text_clean = [x for x in text_lemmetized if len(x)> 0]
    return text_clean
```

## Count Vectorizer

```python
from sklearn.feature_extraction.text import CountVectorizer

# creating instance of count vectorizer and giving it our cleaning function to analyze each data item
count_vect = CountVectorizer(analyzer = cleanData)
word_counts = count_vect.fit_transform(data['text'])
print(word_counts.shape)
```
